Conditionals are used to compare items (comparison operators)
  - ">" : greater than sign
  - ">=" : greater than or equals sign
  - "<" : less than sign
  - "<=" : less than or equals sign
  - "==" : equals too (aka equality)
  - "!=" : not equals too 


**Boolean** expressions result in an answer of true or false, yes or no (only two outcomes to the question) (a binary expression)


*In python arguements get passed to functions as objects

**Match-Case** can be used similar to a switch statement in java, but python notebooks which are powered by Jupyter do not work



```python
def main():

#if, elif, else conditional statements
  x = int(input("what is x? "))
  y = int(input("what is y? "))

  if x < y:
    print ("x is less than y")

  elif x > y:
    print ("x is greater than y")
  else:
    print ("x is equal to y")


# or & and operators
  grade = int(input("what is your grade? "))

  if grade <= 100 and grade >= 90:
    print ("A")

#another way to use the and operator
  elif 80 <= grade < 90:
    print ("B")

  else:
    print ("you need a b to pass and you didn't achieve that ")


#a single line way of doing if else statement
  print ("x is equal to y") if (x == y) else print ("x is not equal to y")



main()
```
