- Libraries are files of code (either created by you or other parties) that you can use in your program 
- A module is a library that has one or more functions in it 
  - Libraries and Modules are great for re-using your own code as well as using other peoples functions so that you can work soley on solving your problem at hand

- import: used to import a module into your program
- from: used to import functions from a module

Command line arguements allow you to pass in values (i.e. arguements) into the program during its initiation 
- sys.argv is a variable that stores the list of words that the user inputted in the command line (sys.argv[0] is the file name and sys.argv[1] stores the first arguement)


Packages are 3rd party libraries that can be import and used in your program
- use https://pypi.org/ to find all python packages
- use the pip command to install packages (pip is a package manager)
  - When you want to use a package in your program you first have to install it via pip install command and then import it at the top of your file


API is an application programing interface (most apis are internet based)
- use the requests package to make web-based calls to online APIs
- can use requests.get(url) to get api values

JSON (java scipt object notation) is a language angnustic format (meaning any programing language can access and read JSON files) used to exchange data between computors






##Random Library

```python
import random #imports entire library
#from random import choice #just imports the choice function from random


#depending on how i import the function will determine the syntax of how I call the choice function
coin = random.choice(["heads", "tails"])
print(coin)

number = random.randint(1,10)
print(number)

cards = ["Jack", "Clubs", "Hearts", "Spades"]
random.shuffle(cards)

for card in cards:
  print(card)
```
