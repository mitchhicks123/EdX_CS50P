Lesson Learnings

---

File Input/Output is all about reading information (loading save info) from or writing information (storing info) to a file
- "open" is a function that can be used to open a file, allowing you to read and write information from/to it
- .txt and .csv are common file extensions to store information to
- .csv (comma seperated values) are common for storing multiple values related to each other (.csv files are also used for when you are exporting data from excel or other software solutions) (.csv files seperate items into rows and colounms using new lines to represent rows and commas to represent colounms)
- example open("filename.txt", "w") (this creates a link to the file for writing)
- "w" : will always create or recreate a new file, brand new with no old data
- "a" : lets you create a file if one doesn't exist, as well as allows you to append new values to it if it already exists

"lamda" functions are functions that are only used in one scenario and therefore can just be written inside of where it would have been traditionally called (lambda functions are known as anonymous functions) 

*Note: general best practice is to  do one action at a time... Ex. if you want to make an edit to a file you first should read through it and store it in memory and then write back to it with the changes (don't try to intertwine unless its an extremely large file)
