The main goal with building tests is simplicity and effectivness

when building a unit test for a program in python you can make another file specifcally for testing and import the file and functions that you want to test as a library

- assert : is a keyword used to test a function by making an assertive claim about what a function may return


pytest is a 3rd party package that can be used to ease the development of unit tests (essentially handles the try and except statements, so you only have to worry about making the unit tests)

Can only run unit tests on functions that have parameters and return values




```python
# this is a dummy file and won't compile, but this is the format for a unit test file without using pytest

import file_to_test

def main():
  test_fileFunctionName()

def test_fileFunctionName():
  try:
    assert fileFunctionName(value) == expected_result
  except AssertionError:
    print("Retun value from file function did not match expected result")




main()


```
