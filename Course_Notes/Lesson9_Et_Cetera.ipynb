Unpacking

````python


def cars(electric, gas, truck):
  print(f"EV: {electric} \nGas: {gas} \nTruck: {truck}")


vehicles = ["tesla", "volvo", "ram"]
#star notation allows you to break up a list while passing it in as a parameter
cars(*vehicles)

print("---------------------------------")

vehicles2 = {"electric": "tesla", "gas": "bmw", "truck": "ford"}
#for dictionaries in order to unpack you need to use two pound signs
cars(**vehicles2)

print("---------------------")

def numbers(*args):
  print(args)

numbers(20,3,18)
numbers(1)

def numbers1(**kwargs):
  print(kwargs)

numbers1(number1=20,number2=3,number3=18)
numbers1(number1= 96)
````

Map / List Comprehensions 

````python
def states(*states):
  upper = map(str.upper, states)
  print(*upper)

def states1(*states):
  upper_list = [states.upper for state in states]
  print(*upper_list)


states("fla", "ca", "mt", "ny")

states("tx", "ut", "ma", "az")
````

Dictionary Comprehension

````python
cars = ["Ram", "Ford", "Tesla", "Jeep"]

north_american = {car: "north american" for car in cars}

print(north_american)
````

Enumerate

````python
cities = ["Toronto", "New York", "Montreal", "Miami", "Los Angels"]


for i, city in enumerate(cities):
  print(i+1, city)

````
