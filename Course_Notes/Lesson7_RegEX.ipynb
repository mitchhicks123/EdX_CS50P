A RegEx is just a pattern used to match on some type of data

RegExs are great for cleaning and sorting inputted data


- re: is a library used for regular expressions. Used to define, set and match patterns. A regular expression is just a pattern
- search function in the "re" module is very powerful
- re.search(pattern, search, flag=0)
- pattern includes items to identify how many characters you want
- search includes the string you would like to search in 
- flag can be used to manipultae the search function

(patterns can include special characters like ". * + ? {m} {n}" see full list here https://docs.python.org/3/library/re.html 

- use the letter r at the start of the string and a backslash ecscape character to type in one of the special characters literally. Example re.search(r".+@.+\.com")

- can use the ":=" operator to effectively assign a value from right to left and evaluate it as a boolean expression for a conditional statement (aka the walrus

- () inside re.search makes groups that can be stored upon completion
