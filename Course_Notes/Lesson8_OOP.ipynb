In a class you have 2 types of methods: instance methods and class methods 
- instance methods that take in at least one arguement "self" (aka reference to it's object) and are used for functionality of an object
- class methods take in one arguement class "cls" and are meant for general functionality where objects either do not exist or don't protain to their functionality

In a class you have 2 types of variables: instance variables and class variables
- instance variables get stored in an object and are created when an object is instaniated (instance methods are the default method)
- class variables get stored in the class and can be access by any method in the class and are available for every object (in order to produce a class method you need to use the property "@classmethod")
