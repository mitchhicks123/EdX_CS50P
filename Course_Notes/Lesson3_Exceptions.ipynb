- Compile time Errors occur when your program is being compiled from the programming language into computer language
- Syntax Errors are created by the programmer and occur when something is inputted inapropriately (happens while the code is compiling)
- Runtime Errors occur when your program is running/being executed 

- Try, Except and Else statements are used to catch errors in python (can think of this as, "try" and execute this code and if you hit an error jump to the "except" statement for when an error occurs, else if everythings fine continue executing to the "else" clause) (try and except clauses should wrap tightly around the code in question)


```python
#try to execute lines 3 and 4, if an error occurs jump to 6 and 7, else if no error occurs jump to lines 9 and 10 and execute the else clause

while True:
  try:
    x = int(input("What's the value of x: "))

  except ValueError:
    print("You must enter an integer")

  else:
    break

print (f"x is {x}")



# another try except example with the use of pass
while True:
  try:
    y = int(input("What's the value of y: "))

  except ValueError:
    pass

  else:
    break

print ("y is " + y)


print ("therefore, the sum of x and y is", x + y)
```
