##Print Functions 

```python

#asks the user for input and stores it in day
day = input ("How's your day? ")


# strip spaces from string and capitalize the string (below is a way to chain methods)
day = day.strip().title()

firstAnswer, secondAnswer = day.split(" ")

"""
multiple ways to concatenate strings
"""
print ("Your day is going " + firstAnswer)
print ("Your day is going", day)

# format string
print (f"Your day is going {day}" )
```

##Math Operations

```python
x = float(input ("Please provide a value for x: "))
y = float(input ("Please provide a value for y: "))

z = round (x + y, 2)

print (z)


#another way to round
q = x / y

print (f"{q:.2f}")

```
